```python
# Приятная библиотека для работы с текстом
"""
The constants defined in this module are:

string.ascii_letters

    The concatenation of the ascii_lowercase and ascii_uppercase constants described below. This value is not locale-dependent.

string.ascii_lowercase

    The lowercase letters 'abcdefghijklmnopqrstuvwxyz'. This value is not locale-dependent and will not change.

string.ascii_uppercase

    The uppercase letters 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'. This value is not locale-dependent and will not change.

string.digits

    The string '0123456789'.

string.hexdigits

    The string '0123456789abcdefABCDEF'.

string.octdigits

    The string '01234567'.

string.punctuation

    String of ASCII characters which are considered punctuation characters in the C locale: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~.

string.printable

    String of ASCII characters which are considered printable. This is a combination of digits, ascii_letters, punctuation, and whitespace.

string.whitespace

    A string containing all ASCII characters that are considered whitespace. This includes the characters space, tab, linefeed, return, formfeed, and vertical tab.
"""
import string

# open => Работа с файлом
# open ({FILENAME}, encoding='{FORMAT}', mode='{MODE}')
# FILENAME -> Путь до файла
# FORMAT ->  Формат чтения для Python, вообще указывать не надо если ENG, а так 'utf-8'
# MODE -> * 'r' => read file
#         * 'w' => write in file
# Остальное в гугле чекаете

file = open('input.txt', encoding='utf-8', mode='r')

# .upper() => Все буквы будут заглавные
# .lower() => Все буквы будут прописные
# .translate() в Python возвращает новую строку, в которой каждый символ в строке заменяется с использованием данной таблицы перевода.
# str.maketrans => Почитайте, но я вам не советую её использовать
# Лучше почитайте о .replace("{BEFORE}", "{AFTER}")
#   * AFTER => Символ, который заменяется
#   * BEFORE => На что заменить символ
text = file.read().upper().translate(str.maketrans('', '', string.punctuation))
# len() => Количество всех символов
symbolLen = len(text)
# {} => Словарь имеющий формат key : value
# _dict.items() => Содержит keys, values
# _dict.values() => Содержит values
# _dict.keys() => Содержит keys
_dict = {}
# set() => Убирает дупликаты
# Например есть массив: a = [1,1,1,1,2,2,2,2,2,2,3,3,3,3,3]
# a = set(a) => a = [1,2,3]
# Циклы for
# for symbol in text, где symbol - каждый символ в тексте
# for value in range(23) - каждое число от 0 до 23(23 не выводится)
# for value in range(1,23) - от 1 до 23(23 не выводится)
# for value in range(1,23,2) - от 1 до 23, где шаг будет +2
for symbol in set(text):
    # Тут каждой уникальной букве(ключ) мы присваем значение 0
    _dict[symbol] = 0

for word in text:
    # Прибавляем единичку к кажой букве, через которую прошел цикл for
    _dict[word] += 1

for symbol in set(text):
    # Узнаем значение относительно количества всех символов
    _dict[symbol] = _dict[symbol]/symbolLen
# Вывод словаря со всеми значениями
print(_dict)
```

```python
#Библиотека для использования мат. функций
import math
def entropy(text):
    _dict = {}
    for symbol in set(text):
        _dict[symbol] = 0

    for word in text:
        _dict[word] += 1

    # Ниже это формула энтропии
    # float => число с плавающей точкой(со знаком после запятой)
    total = len(text)
    #Пояснение
    """
    Цикл формата
    sum = 0
    for symbol in _dict:
        # Допустим так можно
        sum += int(symbol)
    
    Когда ты гигачад то можно писать формата "for loop one line"
    sum = 0
    sum += int(symbol) for symbol in _dict
    """
    probs = [float(_dict[char]) / total for char in _dict]

    entropy = -sum([p * math.log(p, 2) for p in probs])
    return entropy

file = open('input.txt', encoding='utf-8', mode='r')
file = file.read()
print(entropy(file))
```
